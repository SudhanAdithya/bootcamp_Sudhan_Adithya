```python

import pandas as pd
import numpy as np

# Load your dataset here
# df = pd.read_csv("your_dataset.csv")

# --- Example synthetic dataset ---
np.random.seed(42)
df = pd.DataFrame({
    "age": np.random.randint(18, 70, 200),
    "income": np.random.normal(60000, 15000, 200),
    "purchases": np.random.poisson(5, 200),
    "spend_score": np.random.uniform(1, 100, 200),
    "date": pd.date_range("2023-01-01", periods=200, freq="D")
})

df.head()
```

```python

# Info
df.info()

# Numeric summary
df.describe()

# Missing values
df.isna().sum()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Histogram - Age
sns.histplot(df["age"], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()

# Histogram - Income
sns.histplot(df["income"], bins=20, kde=True)
plt.title("Income Distribution")
plt.show()

# Boxplot - Spend Score
sns.boxplot(x=df["spend_score"])
plt.title("Spend Score Boxplot")
plt.show()
```

```python

# Scatter plot: Income vs Purchases
sns.scatterplot(x="income", y="purchases", data=df)
plt.title("Income vs Purchases")
plt.show()

# Line plot: Spend Score over Time
sns.lineplot(x="date", y="spend_score", data=df)
plt.title("Spend Score Over Time")
plt.show()
```

```python

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
